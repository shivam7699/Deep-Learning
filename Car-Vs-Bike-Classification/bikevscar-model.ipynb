```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
data_gen = ImageDataGenerator(zoom_range=0.2, horizontal_flip=True,
                             rotation_range=0.2,rescale=1/255,
                             shear_range=0.05,
                             vertical_flip=True,
                             validation_split=0.20)
```

```python
path_gen = '/kaggle/input/car-vs-bike-classification-dataset/Car-Bike-Dataset'
```

```python
os.listdir('/kaggle/input/car-vs-bike-classification-dataset/Car-Bike-Dataset')
```

```python
train_data = data_gen.flow_from_directory(directory=path_gen,target_size=(244,244),
                                         batch_size=64)
```

```python
train_data.class_indices
```

```python
from tensorflow.keras.layers import Dense,Conv2D,MaxPool2D,Flatten
from tensorflow.keras.models import Sequential
```

```python
model = Sequential()
model.add(Conv2D(32, (3,3), input_shape=(224,224,3)))
model.add(MaxPool2D())
model.add(Conv2D(64, (3,3)))
model.add(MaxPool2D())
model.add(Conv2D(128, (3,3)))
model.add(MaxPool2D())
model.add(Conv2D(128, (3,3)))
model.add(MaxPool2D())
model.add(Conv2D(256, (3,3)))
model.add(MaxPool2D())

model.add(Flatten())
model.add(Dense(2,activation='softmax'))

model.compile(loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(train_data, epochs=10)
```

```python
model.save('bikeVscar.h5')
```
